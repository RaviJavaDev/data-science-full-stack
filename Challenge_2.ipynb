```python
import re
import os
import string
import logging

class FlieExplorer:
    def __init__(self):
        os.chdir('C:\\\Ravis\\Data Science\\Full Stack DS\\Python\\Practice')
        os.makedirs('Challenge_2',exist_ok=True)
        os.makedirs('Challenge_2\\log',exist_ok=True)
        logging.basicConfig(filename=os.path.join(os.getcwd(),'Challenge_2','log','log.txt'),
                            level=logging.DEBUG,format='%(asctime)s %(levelname)s %(message)s')
        
    def get_drive(self):
        logging.info('in get_drive()')
        try:
            drive_list = ['%s:' % d for d in string.ascii_uppercase if os.path.exists('%s:' % d)]
            logging.debug('drives found {}'.format(drive_list))
        except Exception as e:
            logging.error('error {} occured in get_drive() while getting drive.'.format(e))
        logging.info('out get_drive()')
        return drive_list

    def search_file(self,file_name):
        print('file_name',file_name)
        logging.info('in search_file()')
        logging.debug('file to search {}'.format(file_name))
        try:
            rex = re.compile(file_name)
            for drive in self.get_drive():
                for root, dir, files in os.walk(drive):
                    if file_name in files:
                        result = rex.search(file_name)
                        if result:
                            logging.debug('file found {}'.format(os.path.join(root,file_name)))
                            yield os.path.join(root,file_name)
        except Exception as e:
            logging.error('error {} occured in search_file() while searching file.'.format(e))
        logging.info('out search_file()')

    def merge_files(self,file_list,file_name):
        logging.info('in merge_files()')
        logging.debug('files to merge {}'.format(file_name))
        merged_file = 'merged_{}'.format(file_name)
        try:
            f = open(file=os.path.join(os.getcwd(),'Challenge_2',merged_file),mode='w')
            for file_nm in file_list:
                f1 = open(file = file_nm,mode='r')
                f.writelines(f1.readlines())
                f1.close()
            f.close()
            logging.debug('{} files merged successfully into {}'.format(file_list,merged_file))
        except Exception as e:
            logging.error('error {} occured in merge_files() while searching file.'.format(e))
        logging.info('out merge_files()')

```
